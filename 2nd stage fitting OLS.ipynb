```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
X = pd.read_csv("C:/DISSERTATION/Data/X.csv", index_col=0)
n = len(X)
cut_points = [fold * n // 5 for fold in range(5)] + [n]
```

```python
model = LinearRegression()
```

```python
for knob in range(9):
    for dataset in range(10):
        scores = pd.read_csv("C:/DISSERTATION/Results/OLS/scores/dataset_"+str(knob)+str(dataset)+".csv", index_col=0)
        results = pd.DataFrame(index=scores.index, columns=scores.columns)
        for score in scores.columns:
            if score=='T':
                results[score] = scores[score]
            else:
                y = scores[score]
                for fold in range(5):
                    X_in = X.drop(range(cut_points[fold], cut_points[fold+1]))
                    y_in = y.drop(range(cut_points[fold], cut_points[fold+1]))
                    X_out = X.iloc[cut_points[fold] : cut_points[fold+1]]
                    y_out = y.iloc[cut_points[fold] : cut_points[fold+1]]
                    model.fit(X_in, y_in)
                    results[score].iloc[y_out.index] = model.predict(X_out)
        results.to_csv("C:/DISSERTATION/Results/OLS/TFUD/dataset_"+str(knob)+str(dataset)+".csv")
```
